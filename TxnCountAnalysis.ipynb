```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_txn = pd.read_csv("C://Users//1022895//Documents//Python//P1//TXN_COUNTS_RAW.csv")
df_ord = pd.read_csv("C://Users//1022895//Documents//Python//P1//ORDER_COUNTS_RAW.csv")
```

```python
df_txn.head()
```

```python
df_ord.head()
```

```python
df_txn_count = df_txn.copy()
df_ord_count = df_ord.copy()

df_txn_count.columns = [str(i).strip("'") for i in df_txn_count]
df_txn_count = df_txn_count.replace(np.nan, 0)
df_ord_count.columns = [str(i).strip("'") for i in df_ord_count]
df_ord_count = df_ord_count.replace(np.nan, 0)


df_txn_c_g_t = df_txn_count.groupby('Day').mean().T.reset_index()
df_ord_c_g_t = df_ord_count.groupby('Day').mean().T.reset_index()
```

```python
df_ord_c_g_t
```

```python
col = ['index','Monday','Tuesday', 'Wednesday','Thursday','Friday' , 'Saturday', 'Sunday']
df_ord_c_g_t = df_ord_c_g_t[col]
col_list = df_ord_c_g_t.columns


col_list = col_list[1:]
col_list
```

```python
txn_plot = df_txn_c_g_t.plot(x = 'index', y = col_list, xlabel = 'Hour',ylabel = 'Txn count',figsize=(9,6))
```

```python
ord_plot = df_ord_c_g_t.plot(x = 'index', y = col_list, xlabel = 'Hour',ylabel = 'Ord count',figsize=(9,6))
```
